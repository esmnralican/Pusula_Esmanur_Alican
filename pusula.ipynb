```python
# Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python

# import dataset
file_path = '/content/side_effect_data 1.xlsx'
data = pd.read_excel(file_path)

# first 5 data in the dataset
data.head(20)
```

#**DATA PREPROCESSING**
## **Description**:
**Weight and Height:** Since these columns are numeric, the median (middle value) is used to fill in the blank values. Since the median is not affected by extremes, it minimizes imbalances in the distribution.

**Categorical Data (Allergy and Chronic Diseases):** Filled using the most common (mode) values. In this way, missing data is filled with the most common value.

**Medication Start/End Dates:** If there is only a medication start date, 60 days has been added as a typical duration of treatment. This ensures that blank dates are filled in realistically.

```python
# Descriptive statistics for numeric columns
data.describe()
```

```python
# fill the missing values in the 'Weight' and 'Height' columns with the median (the median is not affected by extreme outliers)
data['Kilo'].fillna(data['Kilo'].median(), inplace=True)
data['Boy'].fillna(data['Boy'].median(), inplace=True)

# fill categorical columns such as 'My allergies' and 'My chronic diseases' with the most frequent value (mode)
data['Alerjilerim'].fillna(data['Alerjilerim'].mode()[0], inplace=True)
data['Kronik Hastaliklarim'].fillna(data['Kronik Hastaliklarim'].mode()[0], inplace=True)

# apply a logical method when filling in the start and end dates of medication
# If the medicine has a start date, we fill in the end date with a typical duration added to the start date
import numpy as np
data['Ilac_Baslangic_Tarihi'] = pd.to_datetime(data['Ilac_Baslangic_Tarihi'])
data['Ilac_Bitis_Tarihi'] = pd.to_datetime(data['Ilac_Bitis_Tarihi'])

# for those without an end date, we add 60 days
data['Ilac_Bitis_Tarihi'].fillna(data['Ilac_Baslangic_Tarihi'] + pd.DateOffset(days=60), inplace=True)

# Check the final situation
data.isnull().sum()
```

```python
# Label Encoding was done to convert categorical data into numeric format
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

# data[data.Uyruk != 'Turkiye'] # All nations are Turkish
data['Uyruk'] = label_encoder.fit_transform(data['Uyruk'])
```

##**Plots**


```python
# Convert 'Ilac_Baslangic_Tarihi' and 'Ilac_Bitis_Tarihi' to datetime
data['Ilac_Baslangic_Tarihi'] = pd.to_datetime(data['Ilac_Baslangic_Tarihi'], errors='coerce')
data['Ilac_Bitis_Tarihi'] = pd.to_datetime(data['Ilac_Bitis_Tarihi'], errors='coerce')

# Calculate the difference between the two dates and create a new column 'Ilac_Kullanim_Suresi'
data['Ilac_Kullanim_Suresi'] = (data['Ilac_Bitis_Tarihi'] - data['Ilac_Baslangic_Tarihi']).dt.days


plt.figure(figsize=(10, 6))
sns.boxplot(x='Yan_Etki', y='Ilac_Kullanim_Suresi', data=data)
plt.xticks(rotation=90)
plt.title('Yan Etkilere Göre İlaç Kullanım Süresi')
plt.show()
```

```python
# Visualizing the relationship between 'Height' and 'Weight' with scatter plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Boy', y='Kilo', data=data)
plt.title('Kilo vs Boy')
plt.show()
```

###**Correlation Maps**

```python
# Scenario 1: The Relationship between Age and Chronic Diseases, In this scenario, we analyze the incidence of chronic diseases as we age.

# To examine the relationship between chronic diseases and age
chronic_diseases = ['Hipertansiyon', 'Diyabet', 'Kalp Hastaliklari', 'KOAH', 'Kemik Erimesi', 'Guatr', 'Kan Hastaliklari']

def check_disease(disease, col):
    return data[col].str.contains(disease, na=False).astype(int)

data_numeric = pd.DataFrame()

# Calculating age
data['Dogum_Tarihi'] = pd.to_datetime(data['Dogum_Tarihi'], errors='coerce')
current_year = datetime.now().year
data['Yas'] = current_year - data['Dogum_Tarihi'].dt.year

# Add 'Yas' column to data_numeric DataFrame
data_numeric['Yas'] = data['Yas']

#calculate the correlation between chronic diseases and age
for disease in chronic_diseases:
    data_numeric[disease] = check_disease(disease, 'Kronik Hastaliklarim')

# calculate the correlation between diseases and age
age_chronic_corr = data_numeric[['Yas'] + chronic_diseases].corr()
print(age_chronic_corr)

# Create a heatmap
sns.heatmap(age_chronic_corr, annot=True, cmap='viridis')

# Set the title
plt.title('Correlation Between Age and Chronic Diseases')

# Show the plot
plt.show()
```

```python
# Scenario 2: The Relationship between Gender and Chronic Diseases

# Ensure 'Cinsiyet' column is in data_numeric
data['Cinsiyet'] = data['Cinsiyet'].map({'Male': 1, 'Female': 2})
data_numeric['Cinsiyet'] = data['Cinsiyet']
print(data["Cinsiyet"])

# data[data['Cinsiyet'].notna()]

# Modify the check_disease function to add disease columns to data
def check_disease(disease, col):
    data[disease] = data[col].str.contains(disease, na=False).astype(int)

# Add the 'chronic_diseases' columns to the 'data' DataFrame
for disease in chronic_diseases:
    check_disease(disease, 'Kronik Hastaliklarim')

# To examine the relationship between gender and chronic diseases
sex_chronic_corr = data[['Cinsiyet'] + chronic_diseases].corr()
print(sex_chronic_corr)

# Create a heatmap
sns.heatmap(sex_chronic_corr, annot=True, cmap='viridis')

# Set the title
plt.title('Correlation Between Gender and Chronic Diseases')

# Show the plot
plt.show()
```

```python
# Scenario 3: Family Chronic Diseases and Personal Chronic Diseases

# Relationship between paternal and maternal chronic diseases and personal illnesses
family_columns = ['Baba Kronik Hastaliklari', 'Anne Kronik Hastaliklari']

for col in family_columns:
    for disease in chronic_diseases:
        data[col + '_' + disease] = check_disease(disease, col)

family_disease_corr = data[chronic_diseases + [col + '_' + disease for col in family_columns for disease in chronic_diseases]].corr()

# Create a heatmap to visualize the correlations
plt.figure(figsize=(12, 8))
sns.heatmap(family_disease_corr, annot=True, cmap='viridis')
plt.title('Correlation Between Family and Personal Chronic Diseases')
plt.show()
```

```python
# Scenario 4: Allergies and Chronic Diseases
# Examination of chronic diseases by allergy
data['Alerji'] = data['Alerjilerim'].notnull().astype(int)

allergy_chronic_corr = data[['Alerji'] + chronic_diseases].corr()
print(allergy_chronic_corr)
# Create a heatmap
sns.heatmap(age_chronic_corr, annot=True, cmap='viridis')

# Set the title
plt.title('Allergies and Chronic Diseases')

# Show the plot
plt.show()
```

##**SUGGESTIONS**

**1. Personalized Treatment Plans**
The data can analyze which side effects medication use produces in people of which age group, gender or health history. This can help doctors create more personalized treatment plans:
- **Example Problem:** The same medicine has different side effects in different individuals.
- **Solution:** Potential side effects can be predicted based on patients' demographics, allergies and chronic diseases before starting medication. This enables safer treatment protocols to be established.


**2. Side Effect Risk Assessment**
The likelihood of medicines producing certain side effects can be analyzed based on personal data. For example, a particular drug may produce serious side effects in individuals with certain chronic diseases. Such analyses can create decision support systems for drug manufacturers and doctors.

- **Example Problem:** Determine whether certain medicines have unexpected side effects.

- **Solution:** Through data analysis, it is possible to identify which drugs cause which side effects most often and to develop alternative drug recommendations for high-risk patients.

**3. Drug Interactions with Allergies and Chronic Diseases**
Allergy and chronic disease information in the data allows analyzing how certain medicines interact with these conditions.These analyses can help to understand in which people with which diseases certain medicines may be more risky.
- **Example Problem:** People with certain chronic diseases may experience increased severity of side effects.
- **Solution:** Analyze which medicines require more attention based on allergies and chronic diseases and use this information in clinical decision support.
